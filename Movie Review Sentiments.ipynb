```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('IMDB Dataset.csv')
```

```python
data.head()
```

```python
data['review'][0]
```

```python
import re
```

```python
pattern = re.compile(r'<.*?>') #Didn't use r'<.*>' as it will lead to ex : <h1>Hello</h1> taking only the outer <>
def clean_htmlTags(text):
    #pattern = re.compile(r'<.*?>') , avoid compiling here again and again
    return re.sub(pattern, '', text)
```

```python
data['review'] = data['review'].apply(clean_htmlTags)
```

```python
data['review'][0]
```

```python
def lower_case(text):
    return text.lower()

data['review'] = data['review'].apply(lower_case)
```

```python
data['review'][0]
```

```python
def remove_splChar(text):
    return re.sub(r"[^\w\s']", ' ', text)
data['review'] = data['review'].apply(remove_splChar)
```

```python
data['review'][0]
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stopWords = set(stopwords.words('english'))
```

```python
def remove_stopwords(text):
    x = []
    for i in text.split():
        if i not in stopWords:
            x.append(i)
    return ' '.join(x)
        
```

```python
data['review'] = data['review'].apply(remove_stopwords)
```

```python
data['review'][0]
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    x = []
    for i in text.split():
        x.append(ps.stem(i))
    return ' '.join(x)
        
```

```python
data['review'] = data['review'].apply(stem)
```

```python
data['review'][0]
```

```python
data = data.iloc[:10000]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000) 
#since the libraries used to make vectore, and vector/matrix comprise of columns(features) , hence to limit the columns to save space we use max_features
#Also it helps to achieve greater accuracy, as the rare words are removed which doesn't help in making model any better
#cv is an object of countvector class
```

```python
X = cv.fit_transform(data['review']).toarray()
#when fit applies to the review data, it gathers all unique words from the whole column (scans all rows at once) and provide indexing to each new word
#then the tranform helps in creating a sparse vector with features/Columns as the words gathered with the,
# rows == reviews
# col == words
#It reviews the review again and put frquency in the respective column {stores count for word present, 0 for which is not present}
# each row stores the frequency for the corresponding column as per the frequency of that word encountered in the current review(row) 

# Here it forms the sparse matrix i.e here only that column is mentioned for any row whose word is present in the review, Hence we need to make it array
#We require a dense matrix to know freq of each word in a review so that model can predict therfore did -> .toarray()
```

```python
#Now the sentiment column must be label encoded
data.rename(columns={'sentiment': 'label'}, inplace=True)
data['label'] = data['label'].map({
    'positive': 1,
    'negative': 0
})
```

```python
y = data.iloc[:, -1].values #: all rows, -1 last column, .values -> converts to numpy array
```

```python
## Note : we always split the data into test and trains so that 
## we have a test data upon which we can check our accuracy vefore directly putting data whose result is unknown
```

```python
#Now we have to split data in train_test 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
#In the Training Phase
# the naive bayes prepares following on basis of the training data:

# =>> discrete data
#Lookup table for the : categorical data, Multinomial data, bernoulli data etc, comprising all sorts of probabilities possible 
# like P(feature_value(one of the many (if many values are there for each feature)) | +ve (or -ve)), 
# this helps in computing the rhs of the naive bayes eqn when required while predicting

# =>> continuous data
# Parameters like alpha, mean, varinace for the columns in case of continuous data like exponential or gaussian data
# like for guassian data, each col/ feature is computed with (mu , sigma2) for both +ve and -ve , which is used to compute P(feature = val | +ve (or -ve))
# using them in the pdf for gaussian model with x = val, while computing the rhs of the naive bayes eqn
```

```python
clf1 = GaussianNB()
clf2 = MultinomialNB()
clf3 = BernoulliNB()
#objects of respective classes
```

```python
clf1.fit(X_train, y_train)
clf2.fit(X_train, y_train)
clf3.fit(X_train, y_train)
#fit means "train"
```

```python
pred1 = clf1.predict(X_test)
pred2 = clf2.predict(X_test)
pred3 = clf3.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Gaussian", accuracy_score(y_test, pred1))
print("Multinomial", accuracy_score(y_test, pred2))
print("Bernoulli", accuracy_score(y_test, pred3))
```
